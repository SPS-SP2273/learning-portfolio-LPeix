```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
# Some data for plotting
x = [0, 1, 2, 3, 4, 5]
y = [0, 4, 8, 12, 16, 20]
y_err = [0.0, 0.5, 1.0, 1.5, 2.0, 2.5]
x_err = .25

plt.errorbar(x, y, 
             xerr=.25, yerr=[0.0, 0.5, 1.0, 1.5, 2.0, 2.5],
             color='darkred', linestyle='dashed',
             marker='o', markerfacecolor='yellow',
             ecolor='black',capsize=5,
             label='Y values with error')
plt.xlabel('x-values')
plt.ylabel('y-values')
plt.title('X vs Y')
plt.grid(alpha=.25)
plt.legend(loc='lower right')
plt.show()
```

# Exercise

```python
x_list = []
for number in range(51):
    x_list.append(number)
x = np.array(x_list)
x = np.linspace(0, 50, 500)
y = (np.e**(-x/10))*np.sin(x)
plt.ylim([-1, 1])
plt.plot(x, y, c='steelblue')
plt.grid(alpha=.25)
plt.title('$y = e^{-x/10}sin(x)$', fontsize=20)
plt.xlabel('x', c='darkorange', fontsize=15)
plt.ylabel('y', c='darkorange', fontsize=15, rotation='horizontal')
plt.show()
```
