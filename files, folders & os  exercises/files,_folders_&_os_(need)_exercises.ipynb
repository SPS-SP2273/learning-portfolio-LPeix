## Task 2

```python
def rename_my_file(old_file_name):
    try:
        two_parts = old_file_name.split('_')
        month_date = two_parts[0].split('-')
        month[month_date[0]]
        name_list = [month[month_date[0]], month_date[1], two_parts[1]]
        new_file_name = '-'.join(name_list)
        return new_file_name
    except:
        print(f'{old_file_name} already exists!')
```

```python
rename_my_file('oct-08_singapore.txt')
```

## Task 4

```python
# create a folder to contain all the txt files
os.mkdir('OSexercise')
```

```python
# move all the txt files into OSexercise folder
for OSfiles in glob.glob('os*\*'):
    ext_check = os.path.splitext(OSfiles)
    if ext_check[1] == '.txt':
        new_path = os.path.join('OSexercise', OSfiles.split('\\')[1])
        shutil.move(OSfiles, new_path)
        print(f'{new_path} moved!')
```

```python
glob.glob('OSexercise\*')
```

```python
for OStxtfiles in glob.glob('OSexercise\*'):
    # split the path, get the file name
    old_txtfiles_name = OStxtfiles.split(os.path.sep)[1]
    # rename the file
    new_txtfiles_name = rename_my_file(old_txtfiles_name)
    new_file_path = os.path.join('OSexercise', new_txtfiles_name)
    os.rename(OStxtfiles, new_file_path)
    print(f'{new_txtfiles_name} formatted!')
```
